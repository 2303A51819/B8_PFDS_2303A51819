```python
import pandas as pd
# Create sample Product dataset
products_data = {
'product_id': [1, 2, 3, 4, 5],
'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam'],
'category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Accessories'],
'price': [1200, 25, 75, 300, 50]
}
products_df = pd.DataFrame(products_data)
# Create sample Sales dataset
sales_data = {
'sale_id': [101, 102, 103, 104, 105, 106],
'product_id': [1, 2, 1, 3, 4, 5],
'quantity': [1, 5, 2, 2, 1, 3],
'sale_date': ['2023-01-15', '2023-01-16', '2023-01-17', '2023-01-17', '2023-01-18', '2023-01-18']
}
sales_df = pd.DataFrame(sales_data)
print("Products DataFrame:")
display(products_df)
print("\nSales DataFrame:")
display(sales_df)
```

```python
# Perform inner merge
inner_merged_df = pd.merge(sales_df, products_df, on='product_id', how='inner')
print("Inner Merged DataFrame:")
display(inner_merged_df)
# Perform outer merge
outer_merged_df = pd.merge(sales_df, products_df, on='product_id', how='outer')
print("\nOuter Merged DataFrame:")
display(outer_merged_df)
```

```python
# Calculate total revenue for each sale
inner_merged_df['total_revenue'] = inner_merged_df['quantity'] * inner_merged_df['price']
# Compute total revenue
total_revenue = inner_merged_df['total_revenue'].sum()
print(f"Total Revenue: ${total_revenue:.2f}")
# Compute average price
average_price = inner_merged_df['price'].mean()
print(f"\nAverage Product Price: ${average_price:.2f}")
# Compute product-wise sales
product_sales = inner_merged_df.groupby('product_name')['total_revenue'].sum().reset_index()
print("\nProduct-wise Sales:")
display(product_sales)
```

```python
# Group by product category and find total revenue per category
category_sales = inner_merged_df.groupby('category')['total_revenue'].sum().reset_index()
print("\nCategory-wise Sales:")
display(category_sales)
# Find the best-selling category
best_selling_category = category_sales.loc[category_sales['total_revenue'].idxmax()]
print(f"\nBest Selling Category:\n")
display(best_selling_category)
```

```python
import pandas as pd
# Create sample Teachers dataset
teachers_data = {
'teacher_id': [101, 102, 103, 104, 105],
'teacher_name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
'department': ['Math', 'Science', 'History', 'Math', 'Science']
}
teachers_df = pd.DataFrame(teachers_data)
# Create sample Courses dataset
courses_data = {
'course_id': [201, 202, 203, 204, 205, 206, 207],
'course_name': ['Calculus', 'Physics', 'World History', 'Linear Algebra', 'Chemistry', 'Algebra', 'Biology'],
'teacher_id': [101, 102, 103, 101, 102, 104, 105],
'credits': [4, 4, 3, 4, 4, 3, 4]
}
courses_df = pd.DataFrame(courses_data)
print("Teachers DataFrame:")
display(teachers_df)
print("\nCourses DataFrame:")
display(courses_df)
```

```python
# Perform inner merge
inner_merged_teachers_courses_df = pd.merge(teachers_df, courses_df, on='teacher_id', how='inner')
print("Inner Merged DataFrame (Teachers and Courses):")
display(inner_merged_teachers_courses_df)
# Perform left merge
left_merged_teachers_courses_df = pd.merge(teachers_df, courses_df, on='teacher_id', how='left')
print("\nLeft Merged DataFrame (Teachers and Courses):")
```

```python
# Compute the number of courses per teacher using the inner merge
courses_per_teacher = inner_merged_teachers_courses_df.groupby('teacher_name')['course_id'].count().reset_index()
courses_per_teacher = courses_per_teacher.rename(columns={'course_id': 'number_of_courses'})
# Calculate the average number of courses per teacher
average_courses_per_teacher = courses_per_teacher['number_of_courses'].mean()
print("Number of courses per teacher:")
display(courses_per_teacher)
print(f"\nAverage number of courses per teacher: {average_courses_per_teacher:.2f}")
```

```python
# Group by department and calculate total workload (sum of credits)
department_workload = inner_merged_teachers_courses_df.groupby('department')['credits'].sum().reset_index()
print("\nTotal workload per department (sum of credits):")
display(department_workload)
```
